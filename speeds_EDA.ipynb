# API Connection to Meteo to pull wind speeds

```python
import requests
import json
from datetime import datetime
```

```python
def igc2df(text, flight_date_str):
    # Extract lines starting with "B" (IGC data lines) and strip whitespace
    lines = [line.strip() for line in text.splitlines() if line.startswith("B")]
    flight_date = datetime.strptime(flight_date_str, "%m/%d/%Y").date()  # Convert flight date string to date
    data = []
    
    # Loop through the IGC lines to extract the data
    for line in lines:
        try:
            # Extract relevant information from the IGC line
            time_str = line[1:7]
            lat_deg = int(line[7:9])
            lat_min = float(line[9:15][:-1])  # Remove direction letter
            lat_dir = line[15]
            lon_deg = int(line[16:18])
            lon_min = float(line[18:24][:-1])  # Remove direction letter
            lon_dir = line[24]
            altitude = int(line[25:29])
            ground_speed = (float(line[29:33]) / 100)*1.94384  # Convert to knots
            
            time_obj = datetime.strptime(time_str, "%H%M%S").time()
            timestamp = datetime.combine(flight_date, time_obj)
            
            # Convert latitude and longitude to decimal format
            latitude = lat_deg + (lat_min / 60000)
            if lat_dir.upper() == 'S':
                latitude = -latitude
            longitude = lon_deg + (lon_min / 60000)
            if lon_dir.upper() == 'W':
                longitude = -longitude
            data.append([timestamp, latitude, longitude, altitude, ground_speed])  # Use m/s here
        except ValueError as e:
            print(f"Error parsing line: {line}")
            print(f"Error: {e}")
            continue
    
    columns = ["timestamp", "latitude", "longitude", "altitude", "ground_speed"]
    df = pd.DataFrame(data, columns=columns)  # Create DataFrame with extracted data
    return df
```

```python
# Function to retrieve weather data from Open-Meteo API at 180m
def get_weather_data(lat, lon, flight_datetime):
    url = "https://archive-api.open-meteo.com/v1/archive"
    params = {
        "latitude": lat,
        "longitude": lon,
        "start_date": flight_datetime.date().isoformat(),
        "end_date": flight_datetime.date().isoformat(),
        "hourly": "windspeed_10m,winddirection_10m",  # Request 80m wind data
        "timezone": "UTC"
    }

    # Request weather data from the API
    response = requests.get(url, params=params, timeout=10)

    if response.status_code != 200:
        print("Error retrieving weather data:", response.text)
        return None
    else:
        weather_data = response.json()
        hourly_times = [datetime.fromisoformat(t) for t in weather_data["hourly"]["time"]]
        time_diffs = [abs((t - flight_datetime).total_seconds()) for t in hourly_times]
        closest_index = time_diffs.index(min(time_diffs))  # Find closest time to flight datetime
        wind_speed = weather_data["hourly"]["windspeed_10m"][closest_index]*1.94384  # Use 180m windspeed
        wind_direction = weather_data["hourly"]["winddirection_10m"][closest_index]  # Use 180m wind direction
        return wind_speed, wind_direction, hourly_times[closest_index]
```

```python
filtered_folder_path = Path("filtered")

wind_speeds = []
wind_directions = []
weather_times = []
ground_speeds = []

# Loop through all files in flights['File'] column
for igc_file_path, flight_date_str, enl_start_time in zip(flights['File'], flights['Date (MM/DD/YYYY)'], flights['ENL_Engine_Run_Start_Times']):
    file_path = filtered_folder_path / igc_file_path
    if file_path.exists():
        print(f"Processing file: {file_path.name}")
        
        text = Path(file_path).read_text()
        flight_date_str = flight_date_str.strftime('%m/%d/%Y')

        # Convert the flight date string from the 'Date (MM/DD/YYYY)' column to datetime
        df_flight = igc2df(text, flight_date_str)
        
        # Extract the actual flight date from the first timestamp
        flight_date = df_flight["timestamp"].iloc[0].date()

        # **Use provided ENL engine start time if available**
        if pd.notna(enl_start_time) and isinstance(enl_start_time, (int, str)) and len(str(enl_start_time)) == 6:
            # Convert ENL start time (HHMMSS) into full datetime
            enl_time_str = f"{flight_date} {str(enl_start_time)[:2]}:{str(enl_start_time)[2:4]}:{str(enl_start_time)[4:]}"
            flight_datetime = pd.to_datetime(enl_time_str)
        else:  
            print(f"Missing or invalid ENL start time for {igc_file_path}, using first timestamp.")
            flight_datetime = df_flight["timestamp"].iloc[0]  # Fallback

        lat = df_flight["latitude"].iloc[0]
        lon = df_flight["longitude"].iloc[0]
        
        # Retrieve weather data for the flight at 180m
        weather_data = get_weather_data(lat, lon, flight_datetime)
        if weather_data:
            wind_speed, wind_direction, closest_time = weather_data
            print(f"Flight Date: {flight_datetime.date()}")
            print(f"Flight Start Time (UTC): {flight_datetime.time()}")
            print(f"Location: Latitude {lat}, Longitude {lon}")
            print(f"Closest Weather Data Time: {closest_time}")
            print(f"Wind Speed (10m): {wind_speed} m/s")  
            print(f"Wind Direction (10m): {wind_direction}°\n")
            
            # Append the weather data to the lists
            wind_speeds.append(wind_speed)
            wind_directions.append(wind_direction)
            weather_times.append(closest_time)
        else:
            # Append None if no weather data found
            wind_speeds.append(None)
            wind_directions.append(None)
            weather_times.append(None)
            
        ground_speed = df_flight["ground_speed"].iloc[0] if "ground_speed" in df_flight.columns else None
        ground_speeds.append(ground_speed)
    else:
        print(f"File not found: {file_path}")

# After processing all flights, add the weather data to the flights DataFrame
flights['Wind_Speed_10m'] = wind_speeds
flights['Wind_Direction_10m'] = wind_directions
flights['Weather_Timestamp'] = weather_times
flights['Ground_Speed'] = ground_speeds
```

```python
def correct_speed(gs, wind_knots, theta_rad):
    """
    Applies the wind correction to a given ground speed (gs, in knots) using the wind vector,
    where theta_rad is the difference (in radians) between the flight course and the wind's toward direction.
    """
    return np.sqrt(gs**2 + wind_knots**2 - 2 * gs * wind_knots * np.cos(theta_rad))
```

```python
def classify_manufacturer(row):
    manu = str(row['Manufacture']).upper()
    gtype = str(row['Gtype']).upper()
    
    if manu.startswith('AS') or 'SCHLEICHER' in manu:
        return 'Schleicher'
    elif manu.startswith('DG') or manu.startswith('LS') or 'DG-FLUGZEUGBAU' in manu or 'DG FLUGZEUGBAU' in manu:
        return 'DG/LS'
    elif 'SCHEMPP' in manu or 'VENTUS' in gtype or 'DISCUSS' in gtype or 'DISCUS' in gtype or 'NIMBUS' in gtype:
        return 'Schempp-Hirth'
    else:
        return 'Other'

flights2['Manufacturer_Group'] = flights2.apply(classify_manufacturer, axis=1)
flights2_new['Manufacturer_Group'] = flights2_new.apply(classify_manufacturer, axis=1)
```
