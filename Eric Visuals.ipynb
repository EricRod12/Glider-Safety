```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# --------------------------
# Define the error data
# --------------------------
first_event_mean = {
    "Launch": ["Turbo", "Turbo", "Turbo", "Self-Launch", "Self-Launch", "Self-Launch"],
    "Metric": ["Min", "Max", "Avg", "Min", "Max", "Avg"],
    "Error_%": [17.63, 13.64, 15.17, 16.62, 15.87, 13.89]
}

all_events_mean = {
    "Launch": ["Turbo", "Turbo", "Turbo", "Self-Launch", "Self-Launch", "Self-Launch"],
    "Metric": ["Min", "Max", "Avg", "Min", "Max", "Avg"],
    "Error_%": [-2.92, -6.38, -5.51, -5.78, -1.54, -6.74]
}

# Create DataFrames for each set
df_first_mean = pd.DataFrame(first_event_mean)
df_all_mean = pd.DataFrame(all_events_mean)

# Add an "Event_Type" column for each
df_first_mean["Event_Type"] = "First Event"
df_all_mean["Event_Type"] = "All Events"

# Concatenate the two DataFrames
df_mean = pd.concat([df_first_mean, df_all_mean], ignore_index=True)

# --------------------------
# Create the charts
# --------------------------
df_turbo = df_mean[df_mean["Launch"] == "Turbo"]
df_selflaunch = df_mean[df_mean["Launch"] == "Self-Launch"]

sns.set(style="whitegrid")

# --------------------------
# Chart 1: Turbo flights
# --------------------------
plt.figure(figsize=(8, 6))
ax1 = sns.barplot(
    data=df_turbo,
    x="Metric",
    y="Error_%",
    hue="Event_Type",
    dodge=True,
    palette="Set2"
)
plt.title("Mean Error % for Turbo Flights")
plt.xlabel("Metric")
plt.ylabel("Mean Error (%)")
plt.legend(title="Event Type")

# Annotate each bar with the error percentage
for patch in ax1.patches:
    # Only label the bars in this chart
    height = patch.get_height()
    # Place the text slightly above the top of the bar
    ax1.text(
        patch.get_x() + patch.get_width()/2,  # x-position
        height + (0.5 if height >= 0 else -0.5),  # y-position offset
        f"{height:.2f}",  # text label
        ha="center",
        va="bottom",
        fontsize=9
    )

plt.tight_layout()
plt.show()

# --------------------------
# Chart 2: Self-Launch flights
# --------------------------
plt.figure(figsize=(8, 6))
ax2 = sns.barplot(
    data=df_selflaunch,
    x="Metric",
    y="Error_%",
    hue="Event_Type",
    dodge=True,
    palette="Set2"
)
plt.title("Mean Error % for Self-Launch Flights")
plt.xlabel("Metric")
plt.ylabel("Mean Error (%)")
plt.legend(title="Event Type")

# Annotate each bar with the error percentage
for patch in ax2.patches:
    height = patch.get_height()
    ax2.text(
        patch.get_x() + patch.get_width()/2,
        height + (0.5 if height >= 0 else -0.5),
        f"{height:.2f}",
        ha="center",
        va="bottom",
        fontsize=9
    )

plt.tight_layout()
plt.show()
```

```python
# Ensure that the categorical columns are of 'category' type.
df["terrain_label_at_engine_run"] = df["terrain_label_at_engine_run"].astype("category")
df["is_circling_during_pre_event_window"] = df["is_circling_during_pre_event_window"].astype("category")
# Ensure outcome is categorical or numeric for labeling.
df["Height_Gain_Indicator"] = df["Height_Gain_Indicator"].astype(int)

# -------------------------------
# Plot 1: Count Plot for Terrain Label vs. Height_Gain_Indicator
# -------------------------------
plt.figure(figsize=(10, 6))
sns.countplot(
    data=df,
    x="terrain_label_at_engine_run",
    hue="Height_Gain_Indicator",
    palette="Set2"
)
plt.title("Terrain Label Distribution by Height_Gain_Indicator")
plt.xlabel("Terrain Label at Engine Run")
plt.ylabel("Count")
plt.legend(title="Height_Gain_Indicator", labels=["Height Loss", "Height Gain"])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# -------------------------------
# Plot 2: Count Plot for Circling Behavior vs. Height_Gain_Indicator
# -------------------------------
plt.figure(figsize=(6, 4))
sns.countplot(
    data=df,
    x="is_circling_during_pre_event_window",
    hue="Height_Gain_Indicator",
    palette="Set3"
)
plt.title("Circling Behavior by Height_Gain_Indicator")
plt.xlabel("Circling During Pre-event Window")
plt.ylabel("Count")
plt.legend(title="Height_Gain_Indicator", labels=["Height Loss", "Height Gain"])
plt.tight_layout()
plt.show()

# -------------------------------
# Plot 3: Stacked Bar Chart for Terrain Label Proportions by Height_Gain_Indicator
# -------------------------------
# Create a crosstab of terrain_label vs. Height_Gain_Indicator, normalized by terrain.
terrain_crosstab = pd.crosstab(df["terrain_label_at_engine_run"], df["Height_Gain_Indicator"], normalize="index")

terrain_crosstab.plot(
    kind="bar",
    stacked=True,
    colormap="viridis",
    figsize=(10, 6)
)
plt.title("Proportion of Height_Gain_Indicator by Terrain Label")
plt.xlabel("Terrain Label at Engine Run")
plt.ylabel("Proportion")
plt.legend(title="Height_Gain_Indicator", labels=["Height Gain", "Height Loss"])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
