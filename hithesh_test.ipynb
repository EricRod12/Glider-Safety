```python
import pandas as pd
from datetime import datetime
from pathlib import Path

def parse_igc_for_time(file_name: str, target_time: str):
    """
    Parses the IGC file for the specified target time and returns the corresponding metrics.

    Args:
    - file_name (str): The name of the IGC file to process.
    - target_time (str): The 6-character time to search for in the IGC file (HHMMSS format).

    Returns:
    - pd.DataFrame: DataFrame with the matching time, latitude, longitude, altitude, and ENL.
    """
    file_path = Path("filtered") / file_name

    try:
        with open(file_path, "r") as f:
            lines = [line.strip() for line in f if line.startswith("B")]  # Parse only 'B' records

        # Parse IGC data
        data = []
        for line in lines:
            time_str = line[1:7]  # 'HHMMSS'
            lat_deg = int(line[7:9])
            lat_min = float(line[9:14])
            lat_dir = line[14]
            lon_deg = int(line[15:18])
            lon_min = float(line[18:23])
            lon_dir = line[23]
            altitude = int(line[25:30])  # GPS altitude
            enl = int(line[30:33]) if len(line) > 32 else None  # ENL if present

            # Convert time to HHMMSS format
            try:
                time_obj = datetime.strptime(time_str, "%H%M%S").time()
                time_formatted = time_obj.strftime("%H%M%S")  # Ensure it's a string in HHMMSS
            except ValueError:
                continue  # Skip invalid entries

            # Convert lat/lon to decimal degrees
            latitude = lat_deg + (lat_min / 60000)
            if lat_dir.upper() == 'S':
                latitude = -latitude

            longitude = lon_deg + (lon_min / 60000)
            if lon_dir.upper() == 'W':
                longitude = -longitude

            data.append([time_formatted, latitude, longitude, altitude, enl])

        # Create DataFrame
        columns = ["Time", "Latitude", "Longitude", "Altitude", "ENL"]
        df = pd.DataFrame(data, columns=columns)

        # Extract metrics for target time
        target_data = df[df["Time"] == target_time]

        if not target_data.empty:
            return target_data
        else:
            return None  # No data found for the target time

    except FileNotFoundError:
        print(f"File {file_name} not found.")
        return None
    except Exception as e:
        print(f"Error processing {file_name}: {e}")
        return None


# Iterate through IGC files in 'filtered' folder
filtered_path = Path("filtered")
results = []

for igc_file in filtered_path.glob("*.igc"):
    file_name = igc_file.name

    # Match the IGC file to the 'File' column in times DataFrame
    time_row = times.loc[times["File"] == file_name]

    if not time_row.empty:
        target_time = str(time_row["ENL_Engine_Run_Start_Times2"].values[0])[:6]  # Extract first 6 characters

        if target_time and target_time.isdigit():
            print(f"Processing {file_name} for time {target_time}...")
            result = parse_igc_for_time(file_name, target_time)

            if result is not None:
                # Add file name for reference
                result.insert(0, "File", file_name)

                # Add additional columns from times DataFrame
                result["terrain_elev_at_first_engine_event (ft)"] = time_row["terrain_elev_at_first_engine_event (ft)"].values[0]
                result["dynamic_AGL_at_first_engine_event (ft)"] = time_row["dynamic_AGL_at_first_engine_event (ft)"].values[0]

                results.append(result)
            else:
                print(f"No data found for {file_name} at time {target_time}")

# Combine all results into a single DataFrame
if results:
    final_df = pd.concat(results, ignore_index=True)
    print(final_df)
else:
    print("No matching data found.")
```

```python
import requests

def fetch_elevation(longitude, latitude):
    url = f"https://epqs.nationalmap.gov/v1/json?x={longitude}&y={latitude}&units=Feet&wkid=4326&includeDate=False"
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an exception for HTTP errors
        data = response.json()
        return data.get("value", None)  # Return the elevation value
    except (requests.RequestException, ValueError) as e:
        print(f"Error fetching elevation data: {e}")
        return None

# Apply the function to each row in the DataFrame
final_df['Terrain_EPQS'] = final_df.apply(
    lambda row: fetch_elevation(row['Longitude'], row['Latitude']), axis=1
)

# Check for missing elevation data
if final_df['Terrain_EPQS'].isnull().any():
    print("Some elevation data could not be retrieved.")

final_df
```

```python
final_df['Terrain_DIFF'] = final_df['terrain_elev_at_first_engine_event (ft)'] - final_df['Terrain_EPQS']
```
